# building and kde Kernel Density Estimation plot. is a non-parameter way to estimate the probablity dendsity function of random variable.
### this kde explains the Normal or Gaussian distribution: in probability theory it's very common continiouse probability distribution.
#### Normal distribution are important in statistics are often used in the natural and social sciences to represent real-valued random variables whose distribution is met.

### just like determining how everyone did the exam, or determining peoples age.

```python
#  don't worry about understanding this code! just for diagram explanations.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# create dataset
dataset = np.random.randn(25)

# creating another ruplot
sns.rugplot(dataset);

# set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x-min to x_max
x_axis = np.linspace(x_min, x_max, 100)

# set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/kernel_density_estimaton#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset))) ** .2

# create an empty kernel list
kernel_list = []

# plot each basic function
for data_point in dataset:
    
    
    #create a kernel for each point and append to list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis, kernel, color = 'grey', alpha = 0.5)
    
plt.ylim(0,1)
```

# let's sum them all up to get the normal distribution basis function

```python
# To get the kde plot we can sum these basis functions.

# plotting the sum of the basis function
sum_of_kde = np.sum(kernel_list, axis=0)

# plot figure
fig = plt.plot(x_axis, sum_of_kde, color='green')

# add the initial rugplot
sns.rugplot(dataset, c = 'red')

# get rid of y-tick marks
plt.yticks([])

# set title
plt.suptitle("Sum of the Basis Funtions")
```
